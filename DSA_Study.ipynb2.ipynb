#고객번호 기준으로 한사람이 몇번 구매,수량은? 성별에 따라 ~



#고객의 구매정보를 활용해 한명의 고객에 대한구매 금액,방문횟수,구매수량 등을 계산하여, 고객의 인구통계적 특성과 결합해 분석

```python
df_product
```

```python
#1. 구매 데이터에서 고객의 구매금액, 방문 횟수, 구매수량과 같은 통계량을 계산
p1=df_product.pivot_table(index='회원번호',values=['구매금액','구매수량'],aggfunc='sum').reset_index()
#pivot_table 함수사용 통계량 계산
#reset_index():Pivot talble의 결과로 index로 잡혀진 데이터를 다시 index를 재정렬해줌
```

한명의 회원번호별로 구매금액과 구매수량의 총합을 하나의 리스트로 나오게 함

```python
p1 #변수로 선언함
```

리스트가 아닌 딕션러리로 컬럼

```python
p2=p1.rename(columns={'구매금액':'총구매금액','구매수량':'총구매수량'})
#딕션러리 중괄호
p2
```

#회원번호와 일치한 애랑 연결
#두데이터 테이블 병합merge함수로 병합할 수 있음(기준컬럼 키컬럼),병합전략=how

```python
#회원번호와 일치한 애랑 연결
#두데이터 테이블 병합merge함수로 병합할 수 있음(기준컬럼 키컬럼),병합전략=how
m1=pd.merge(df_member, p2, on='회원번호', how='inner')
#how에서 inner 공통된 교집합,outerjoin 모두 합집합 없는 수는 결측값
#Right/Left join 두테이블에서 원하는 방향으로만 추출하고 없는건 결측값

m1
```

```python
m1.head(2)
```

1년간 총 한 고객이 방문횟수
한사람이 하루에 똑같은 걸 여러개 살 수 있기에
한명의 공급일자에 1이라고 되게 처리를 해줄거임
날짜별로

```python
df_product['Label']=1
#한명의 회원이 몇번 방문했는가 계산
p3=df_product.pivot_table(index=['회원번호','공급일자'],values='Label',aggfunc='min').reset_index()
#min하면 최소값이 이기에 이 고객이 온날에만 1로 정함
p3
```

```python
p4=p3.pivot_table(index='회원번호', values='Label', aggfunc='sum').reset_index()
p5=p4.rename(columns={'Label':'총방문횟수'})
p5
```

#방문 횟수 계산한 p5데이터와 기존 병합된 m1데이터를 병합

```python
df2 = pd.merge(m1, p5, on='회원번호', how='inner')
df2.head(2)
```

#순서형,시간,날짜에 따른 연속형 데이터의 추세확인 경우(x순서형/Y연속형)


```python
df1['결제시각']
```

```python
#split():특정 문자를 기준으로 데이터를 분할 하는 함수
def func1(x):
  return x.split(" ")

df1['결제시각'].apply(func1)
```

```python
#split():특정 문자를 기준으로 데이터를 분할 하는 함수
def func1(x):
  return x.split(" ")[0] #0번째 날짜데이터만 가져올거다

df1['결제시각'].apply(func1) #실제 데이터는 어렵기에 전처리 하는 법을 알아야함
```

```python
df1['결제일자']=df1['결제시각'].apply(func1) #실제 데이터는 어렵기에 전처리 하는 법을 알아야함
```

```python
sns.lineplot(data=df1,x='결제일자',y='최종결제금액') #seabron의 lineplot사용
```
