```python
#데이터 필더 (특정 조건을 만조가는 데이터를 추출)
cond1=(df1["정지시간2"]>=600)
df1.loc[cond1]
```

```python
#두조건을 모두 만족하는 데이터만 추출
cond1=(df1["정지시간2"]>=600)
df1.loc[cond1]
cond2=(df1["작업구분"]=="야간")
df1.loc[cond1&cond2] #합침 두조건을 모두 만족한 데이터를 추출
```

```python
#두조건중 하나라도 만족하면 추출
cond1=(df1["정지시간2"]>=600)
df1.loc[cond1]
cond2=(df1["메세지"]=="야간")
df1.loc[cond1|cond2] #역슬레시 달러표시
```
