```python
Y=df1["Target"] #water 내 불량 chip개수
X=df1[["Temp_OXid","Oxid_time","thickness",'Line_CD',"Etching_rate",'Temp_Etching']]
```

```python
#학습데이터와 검증데이터를 분할
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1234)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#선형회귀
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#평가
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#예측값 먼저 계산 학습데이터와 검증데이터를각각 계산 이제 예측값과 실제값을 계산
Y_train_pred=model.predict(X_train)
Y_test_pred=model.predict(X_test)
```

```python
r2_score(Y_train,Y_train_pred) #r스쿼어 값 학습데이터
```

```python
r2_score(Y_test,Y_test_pred) #r스쿼어 값 검증데이터
```

```python
mean_squared_error(Y_train,Y_train_pred) #mse 학습데이터
```

```python
mean_squared_error(Y_test,Y_test_pred) #mse 검증데이터mean_absolute_error
```

```python
mean_absolute_error(Y_train,Y_train_pred) #mae 학습데이터
```

```python
mean_absolute_error(Y_test,Y_test_pred) #mae 검증데이터
```

```python
#한번에 쓰기
def evaluation_reg(model):
    #예측값 먼저 계산 학습:데이터와 검증데이터를각각 계산 이제 예측값과 실제값을 계산
  Y_train_pred=model.predict(X_train)
  Y_test_pred=model.predict(X_test)
  print("학습데이터 결정계수:",r2_score(Y_train,Y_train_pred))
  print("검증데이터 결정계수:",r2_score(Y_test,Y_test_pred))
  print("학습데이터 MSE:",mean_squared_error(Y_train,Y_train_pred))
  print("검증데이터 MSE:",mean_squared_error(Y_test,Y_test_pred))
  print("학습데이터 MAE:",mean_absolute_error(Y_train,Y_train_pred))
  print("검증데이터 MAE:",mean_absolute_error(Y_test,Y_test_pred))
```

```python
#과적합
#학습데이터에 대해서 성능이 높게 나오지만, 검증데이터에 서능이 매우저조하게 나오는 현상
```

```python
Y=df1["Target"]
X=df1[['Oxid_time', 'thickness', 'Vapor','resist_target','Line_CD', 'Etching_Chamber', 'UV_type']]
#X에 숫자랑 문자 같이->분리할수 있음
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1234)
```

```python
#문자데이터와 숫자데이터를 구분하여 처리하는 프로세스를 구축할 때,사용하는 함수
from sklearn.compose import ColumnTransformer

#특성공학+학습
from sklearn.pipeline import make_pipeline

#Scaling&Enconding
from sklearn.preprocessing import MinMaxScaler


from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import make_column_transformer

```

#연속형 데이터 이름을 리스트로 선
왜냐면 숫자랑 문자 구분해야하니까

```python
#숫자 해당하는리스트
numeric_list=['Oxid_time', 'thickness','resist_target','Line_CD', 'Etching_Chamber']
category_list=["Vapor","UV_type"]
```

```python
#연속형 데이터가 처리되는 process 결측값으로 평균으로 대치한 뒤, Scale를 최소0/최대1 로 변환
numeric_pipe=make_pipeline(SimpleImputer(strategy="mean"),MinMaxScaler())
numeric_pipe
```

```python
#범주형 데이터가 처리되는 process pipe
#결측값을 최빈값으로 대치한뒤, 문자데이터를 oneHoy인코딩 변환
category_pipe=make_pipeline(SimpleImputer(strategy="most_frequent"),OneHotEncoder())
category_pipe
```

```python
preprocessing_pipe=make_column_transformer((numeric_list,numeric_pipe),
 (category_list,category_pipe))
preprocessing_pipe
#연속형과 범주형을 나누어 처리하는 pipe를 구성함
```

```python
pipe_model=make_pipeline(preprocessing_pipe,LinearRegression())
pipe_model
#선형회귀를 해주는 파이프 모델을 만들어 준것 휴~
```

```python
#이제학습시켜줌
pipe_model.fit(X_train,Y_train)
```

```python
evaluation_reg(pipe_model)
```

 교차 검증 Cross validation (CV)

```python
from sklearn.model_selection import GridSearchCV
```

```python
#5회교차검증을 실시하며 학습을 수행(Pice Model의 Process를 따라가며)
```

```python
grid_model=GridSearchCV(pipe_model,param_grid={},cv=5)
grid_model.fit(X_train,Y_train)
```

캡처.PNG

```python
#교차 검증 이후 가능 성능이 좋은 MOdel선택
best_model=grid_model.best_estimator
```

```python
evaluation_reg(best_model)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#연속형과 범주형을 나누어 처리하는 pipe를 구성
preprocess=make_column_transformer((numeric_pipe,numeric_list),(category_pipe,category_list))
pipe_model=make_pipeline(preprocess,DecisionTreeRegressor())
```

```python
#depth:5~10 5개/split:1~10/leaf 5~10 5개->250가지 경우의 수
hyperparamter={'decisiontreeregressor_max_depth':range(5,11),
               'decisiontreeregressor_min_samples_depth':range(1,11),
               'decisiontreeregressor_min_samples_leaf':range(5,11)}

grid_model=GridSearchCV(pipe_model,param_grid=hyperparamter,cv=5)
#5회의 교차검증 수행
grid_model.fit(X_train,Y_train)
```

```python
best_model=grid_model.best_estimator_
best_model
```

```python
evaluation_reg(best_model)
```
