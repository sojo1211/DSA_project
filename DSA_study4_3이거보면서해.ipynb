#특성 공학+학습 pipe구성


```python
#연속형 데이터와 범주형 데이터로 따로 처리하는 pipe형태
from sklearn.compose import make_column_transformer

#각각의 처리기법을 파이트 형태로 구성
from imblearn.pipeline import make_pipeline #imblearn을 추가해주기
```

```python
#연속형 데이터의 전처리(결측값)
from sklearn.impute import KNNImputer

from sklearn.preprocessing import MinMaxScaler #정규화 스케일작업
```

```python

#범주형 데이터 전처리
from sklearn.impute import SimpleImputer
#단순대치 문자열 처리를 할떄 최빈값으로처리

from sklearn.preprocessing import OneHotEncoder
#원핫인코딩 처리 문자 데이터를 숫자데이터로 변환
```

```python
#imblanced Data Sampling
#비율이 깨져있음 학습전단계에 비율을 맞춰주는
from imblearn.over_sampling import SMOTE
#페이크 데이터를 뿔러서 더 학습시킴

#학습 알고리즘으로 학습
from sklearn.tree import DecisionTreeClassifier
#분류모델 회귀인지 아닌지

#교차검증+Hyperparamter Tuning
from sklearn.model_selection import GridSearchCV

#평가지표
from sklearn.metrics import classification_report
```

```python
from sklearn.model_selection import train_test_split

# 학습데이터와 검증데이터 분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1234)
```

```python
#숫자데이터 프로세스,문자데이터 프로세스
numeric_list=['resist_target','Line_CD','Etching_rate','thickness','Temp_Etching']
category_list=['Vapor','UV_type']
```

```python
 numeric_pipe=make_pipeline(KNNImputer(),MinMaxScaler()) #knn통과 시키고->minmax 통과시킴
 category_pipe=make_pipeline(SimpleImputer(strategy='most_frequent'),OneHotEncoder())
 #결측치가 있다면 최빈값으로 바꿔줌
```

```python
preprocessing_pipe=make_column_transformer((numeric_pipe,numeric_list),(category_pipe,category_list))
#컴포즈를 이용해 결합
```

```python
pipe_model=make_pipeline(preprocessing_pipe,SMOTE(random_state=1234),DecisionTreeClassifier(random_state=1234))
#smote로 비율을 맞춤 우리가 최종적으로 만든 파이프라인->,DecisionTreeClassifier로 학습수행
```

```python
pipe_model
```

```python
hyerparameter = {
    'decisiontreeclassifier__max_depth': range(5, 11),
    'decisiontreeclassifier__min_samples_split': range(5, 11),
    'decisiontreeclassifier__min_samples_leaf': range(5, 11)
} #여기서 알고리즘 변경

# 하이퍼파라미터 튜닝
grid_model = GridSearchCV(pipe_model, param_grid=hyerparameter, cv=3, n_jobs=-1, scoring='f1')

grid_model.fit(X_train, Y_train)
```

```python
best_model=grid_model.best_estimator_
```

```python
def evaluation_cla(model):
  Y_train_pred=model.predict(X_train)
  Y_test_pred=model.predict(X_test)

  print('학습능력 평가:')
  print(classification_report(Y_train,Y_train_pred))
  print('일반화 능력 평가:')
  print(classification_report(Y_test,Y_test_pred))
```

```python
evaluation_cla(best_model)
```
