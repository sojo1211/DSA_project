#학습 데이터와 검증 데이터 분할
Y=df1['target_binom']
#drop() 데이터 프레임에서 불필요한 항목을 제거
X=df1.drop(columns=['Target', 'Error_message', 'target_binom', 'Chamber_Route'
                    ,'Unnamed: 0', 'Ox_Chamber', 'process', 'type','process 2']) #colums에서 있는 값들을 리스트를 써서 제거

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=1234)
#학습 데이터와 검증데이터를 8:2비율 분할 이기에
#randow_state:무작위 추출을 원하는 index로 맞추어 추출 고정시키는 거지이제
print(X_train.shape)
print(X_test.shape)
```

```python
#학습수행
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,Y_train) #fit학습수행
```

```python
#결측값 (Missing Value)
#제거:결측값있는 행을 삭제
#2) 대치:다른 값을 결측값에 대치(평균/중앙값)->알고리즘(KNN)
```

```python
model
```

```python
#평가
from sklearn.metrics import accuracy_score

#w정확도:전체 데이터 중에 정확하게 분류한 데이터의 수
Y_train_pred=model.predict(X_train) #학습능력 예측값Y_train_pred
accuracy_score(Y_train,Y_train_pred) #학습한 실제량과 학습한 예측량
```

```python
#테스트셋 학습 안번도 사용안된 것
Y_test_pred=model.predict(X_test) #학습능력 예측값Y_train_pred
accuracy_score(Y_test,Y_train_pred) #학습한 실제량과 학습한 예측량
#0.867~값이 나와야함 왜 틀린지 모르겠음
```
