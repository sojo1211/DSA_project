1. 자료구조 Dictionary를 이용하여, if 문 없이 연도를 입력했을 때 윤년/평년을 판별하여 출력하
는 프로그램을 생성하시오.

4로 나누어 떨어지는 해:윤년
/평년은 365일

```python
# 윤년과 평년 판별을 위한 딕셔너리
leap_years = {
    True: "윤년",
    False: "평년"
}

# 사용자로부터 연도 입력받기
year = int(input("연도를 입력하세요: "))

# 윤년 여부 판별 (윤년이면 True, 아닐 경우 False)
is_leap_year = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# 결과 출력
print(leap_years[is_leap_year])
```

4. For 반복문을 이용하여, 1부터 입력받은 수 까지 제곱 합을 구하는 프로그램을 만드시오.
Ex ) 1² + 2² + 3² + 4² + ... + n²


```python
# 사용자로부터 입력받기
n = int(input("숫자를 입력하세요: "))

# 제곱 합 계산
제곱합 = 0
for i in range(1, n + 1):
    제곱합 += i ** 2  # i의 제곱을 제곱합에 더함

# 결과 출력
print(f'1² + 2² + ... + {n}² = {제곱합}')
```

6. 데이터에서 고객들의 유입경로(Channel)을 분석하고자 한다. 유입경로(Channel)중 가장 많은
고객이 유입된 경로를 확인하시오.


value_counts()는 pandas의 Series 메서드 중 하나로, 특정 열(Series)에서 각 고유 값의 출현 빈도를 계산하여 반환

```python
Channel_counts = df1['Channel'].value_counts() #pandas의Series에서vaule_caunts()특정열 빈도 수 구할수있음

# 가장 많은 고객이 유입된 경로 확인
가장_많은_유입경로 = Channel_counts.idxmax()  # 가장 많이 유입된 경로,idxmax()기능: Series에서 가장 큰 값의 인덱스를 반환합니다.
가장_많은_고객수 = Channel_counts.max()     # 해당 경로의 고객 수 max()기능: Series에서 가장 큰 값을 반환합니다.용도: 특정 열의 최대값

print(f'가장 많은 고객이 유입된 경로: {가장_많은_유입경로}, 고객 수: {가장_많은_고객수}')
```

11. 제품군(Product_Type)중 빈도수가 가장 높은 상위 3개의 항목을 추출하여, df2로 선언한 뒤,
계약일자(Datetime)에 따른 월렌탈비용(Amount_Month)의 합을 계산해 result5.xlsx파일로 저
장하시오.


```python
df2=df1.sort_values(by='Product_Type',ascending=False).head(3)
df2
```

```python
result = df2.groupby('Datetime')['Amount_Month'].sum().reset_index()

# 결과를 Excel 파일로 저장

result.to_excel('result5.xlsx', index=False)
files.download('result5.xlsx')
```

df2.groupby('Datetime'):
df2 DataFrame을 Datetime 열의 값에 따라 그룹화합니다. 즉, 동일한 날짜를 가진 행들을 하나의 그룹으로 묶습니다.


['Amount_Month']:
그룹화된 결과에서 Amount_Month 열을 선택합니다. 이 열은 월렌탈비용을 나타내며, 그룹화된 각 날짜에 대해 이 값을 사용할 것입니다.

.sum():
선택된 Amount_Month 열의 값들을 그룹별로 합산합니다. 즉, 각 날짜별로 월렌탈비용의 총합을 계산합니다.

.reset_index():
그룹화된 결과는 기본적으로 인덱스가 그룹화된 키(여기서는 Datetime)로 설정됩니다. reset_index()를 사용하면 이 인덱스를 기본 숫자 인덱스로 되돌리고, 그룹화된 키를 다시 DataFrame의 열로 추가합니다. 이렇게 하면 결과가 더 쉽게 읽을 수 있는 형태가 됩니다.
