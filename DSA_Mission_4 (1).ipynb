6. '요일'에서 '주말'과'주중'을 구분하여,'주중과 주말의 구매금의 평균의 차이가 있는지 가설
검정을 수행

```python
df_product.value_counts()
```

```python
import scipy.stats as stats
df_product['요일'] #주중은 월요일부터 금요일까지, 주말은 토요일과 일요일을 말합니다
cod1=df_product[df_product['요일'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]
cod2=df_product[df_product['요일'].isin(['Saturday', 'Sunday'])]

# '구매금액' 열 선택
cod1_purchase_amount = cod1['구매금액']
cod2_purchase_amount = cod2['구매금액']
# 평균 계산
mean_weekday = cod1_purchase_amount.mean()
mean_weekend = cod2_purchase_amount.mean()


# 독립된 두 표본 t-검정 수행
t_stat, p_value = stats.ttest_ind(mean_weekday, mean_weekend)

# 결과 출력
print(f"주중 구매금액 평균: {mean_weekday}")
print(f"주말 구매금액 평균: {mean_weekend}")
print(f"t-통계량: {t_stat}, p-값: {p_value}")

# 유의수준 설정 (예: 0.05)
alpha = 0.05
if p_value < alpha:
    print("주중과 주말의 구매금액 평균 차이가 통계적으로 유의미합니다.")
else:
    print("주중과 주말의 구매금액 평균 차이가 통계적으로 유의미하지 않습니다.")
```

12. '1회방문구매평균'을 바탕으로 고객등급을 분류하고자 한다. 10만원 이상 고객은 A / 5만
원 이상은 B / 2만원 이상 C / 나머지는 D 등급으로 분류하여, '고객등급'이라는 새로운 항목
을 생성


```python
df1.columns
```

```python
import pandas as pd


df1['1회방문구매평균'] = df1['구매금액'] / df1['방문횟수']

# 고객등급 분류 함수
def classify_customer(row):
    if row['1회방문구매평균'] >= 100000:
        return 'A'
    elif row['1회방문구매평균'] >= 50000:
        return 'B'
    elif row['1회방문구매평균'] >= 20000:
        return 'C'
    else:
        return 'D'

# '고객등급' 열 생성
df1['고객등급'] = df1.apply(classify_customer, axis=1)

# 결과 확인
print(df1[['구매금액', '방문횟수', '1회방문구매평균', '고객등급']])
```
